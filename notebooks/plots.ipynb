```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DNN_train_R2 = np.load("metrics/DNN_train_R2.npy")
DNN_valid_R2 = np.load("metrics/DNN_valid_R2.npy")
DNN_test_R2 = np.load("metrics/DNN_test_R2.npy")

DNN_T_train_R2 = np.load("metrics/DNN-T_train_R2.npy")
DNN_T_valid_R2 = np.load("metrics/DNN-T_valid_R2.npy")
DNN_T_test_R2 = np.load("metrics/DNN-T_test_R2.npy")

SFNP_train_R2 = np.load("metrics/SFNP_train_R2.npy")
SFNP_valid_R2 = np.load("metrics/SFNP_valid_R2.npy")
SFNP_test_R2 = np.load("metrics/SFNP_test_R2.npy")

SFNP_T_train_R2 = np.load("metrics/SFNP-T_train_R2.npy")
SFNP_T_valid_R2 = np.load("metrics/SFNP-T_valid_R2.npy")
SFNP_T_test_R2 = np.load("metrics/SFNP-T_test_R2.npy")

Transformer_train_R2 = np.load("metrics/Transformer_train_R2.npy")
Transformer_valid_R2 = np.load("metrics/Transformer_valid_R2.npy")
Transformer_test_R2 = np.load("metrics/Transformer_test_R2.npy")

Transformer_T_train_R2 = np.load("metrics/Transformer-T_train_R2.npy")
Transformer_T_valid_R2 = np.load("metrics/Transformer-T_valid_R2.npy")
Transformer_T_test_R2 = np.load("metrics/Transformer-T_test_R2.npy")
```

```python
n = np.arange(26)

plt.plot(n, DNN_train_R2, label="DNN")
# plt.plot(n, DNN_T_train_R2, label="DNN-T")
plt.plot(n, SFNP_train_R2, label="SFNP")
plt.plot(n, SFNP_T_train_R2, label="SFNP-T")
plt.plot(n, Transformer_train_R2, label="Transformer")
plt.plot(n, Transformer_T_train_R2, label="Transformer-T")
plt.legend()
plt.title("Train R2")
plt.show()

plt.plot(n, DNN_valid_R2, label="DNN")
# plt.plot(n, DNN_T_valid_R2, label="DNN-T")
plt.plot(n, SFNP_valid_R2, label="SFNP")
plt.plot(n, SFNP_T_valid_R2, label="SFNP-T")
plt.plot(n, Transformer_valid_R2, label="Transformer")
plt.plot(n, Transformer_T_valid_R2, label="Transformer-T")
plt.legend()
plt.title("Valid R2")
plt.show()

plt.plot(n, DNN_test_R2, label="DNN")
# plt.plot(n, DNN_T_test_R2, label="DNN-T")
plt.plot(n, SFNP_test_R2, label="SFNP")
plt.plot(n, SFNP_T_test_R2, label="SFNP-T")
plt.plot(n, Transformer_test_R2, label="Transformer")
plt.plot(n, Transformer_T_test_R2, label="Transformer-T")
plt.legend()
plt.title("Test R2")
plt.show()
```
